```python
time_period = ('1982-01', '2100-12')
```

```python
ds_sia_LENS2 = ds_sia_LENS2.sel(time=slice(*time_period))
ds_sia_LENS1 = ds_sia_LENS1.sel(time=slice(*time_period))
```

```python
# --- Step 0: Compute 1982–2002 baseline climatology ---
clim_LENS2 = ds_LENS2.sel(time=slice("1982", "2002")).mean("time")
clim_LENS1 = ds_LENS1.sel(time=slice("1982", "2002")).mean("time")

clim_qp_LENS2 = ds_q_prime_LENS2.sel(time=slice("1982", "2002")).mean("time")
clim_qp_LENS1 = ds_q_prime_LENS1.sel(time=slice("1982", "2002")).mean("time")

# --- Convert full datasets into anomalies w.r.t 1982–2002 ---
ds_new_LENS2_anom = ds_LENS2 - clim_LENS2
ds_new_LENS1_anom = ds_LENS1 - clim_LENS1

ds_qprime_LENS2_anom = ds_q_prime_LENS2 - clim_qp_LENS2
ds_qprime_LENS1_anom = ds_q_prime_LENS1 - clim_qp_LENS1
```

### LENS2 contributions

```python
# --- Compute contributions per time, ensemble ---
ice_albedo = ds_new_LENS2_anom['II'].sel(time=slice(*time_period))
ice_area   = ds_new_LENS2_anom['I'].sel(time=slice(*time_period))
incoming_SW = (ds_new_LENS2_anom['IV'] + ds_new_LENS2_anom['V']).sel(time=slice(*time_period))

SW_ice = ds_qprime_LENS2_anom['q_prime_ice'].sel(time=slice(*time_period))
SW_ocn = ds_qprime_LENS2_anom['q_prime_ocn'].sel(time=slice(*time_period))

contrib_total = ice_albedo + ice_area + incoming_SW
contrib_ice = ice_albedo + SW_ice
contrib_ocn = ice_area + SW_ocn
```

```python
month = 5,6,7,8

contrib_total_MJJA = (contrib_total.sel(time=contrib_total['time'].dt.month.isin([month]))).resample(time="1Y").mean()
contrib_ice_MJJA = (contrib_ice.sel(time=contrib_ice['time'].dt.month.isin([month]))).resample(time="1Y").mean()
contrib_ocn_MJJA = (contrib_ocn.sel(time=contrib_ocn['time'].dt.month.isin([month]))).resample(time="1Y").mean()

# # just term I
contrib_ice_area_MJJA = (ice_area.sel(time=ice_area['time'].dt.month.isin([month]))).resample(time="1Y").mean()
contrib_ice_albedo_MJJA = (ice_albedo.sel(time=ice_albedo['time'].dt.month.isin([month]))).resample(time="1Y").mean()
contrib_sw_MJJA = (incoming_SW.sel(time=incoming_SW['time'].dt.month.isin([month]))).resample(time="1Y").mean()

```

```python
sia_PIP = (ds_sia_LENS2['sia_may'].dropna(dim='time').values +\
           ds_sia_LENS2['sia_jun'].dropna(dim='time').values+\
           ds_sia_LENS2['sia_jul'].dropna(dim='time').values+\
           ds_sia_LENS2['sia_aug'].dropna(dim='time').values)*0.25
```

```python
time = np.arange("1982", "2101", dtype="datetime64[Y]")

# Infer ensemble size from NumPy arrays
ensemble_size = 100  # Assuming first axis is ensemble

# Create xarray DataArrays
sia_PIP_xr = xr.DataArray(sia_PIP, dims=("ensemble", "time"), coords={"ensemble": np.arange(ensemble_size), "time": time})

# Combine into a dataset
ds_sia_avg = xr.Dataset({
    "sia_PIP": sia_PIP_xr
})
```

Figure for LENS2

```python
fig, ax = plt.subplots(figsize=(6,6))



# Scatter all ensemble points (optional)
for ens in ds_sia_avg.ensemble:
    ax.scatter(1e-16*ds_sia_avg['sia_PIP'].sel(ensemble=ens), contrib_ice_area_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Blues')
    ax.scatter(1e-16*ds_sia_avg['sia_PIP'].sel(ensemble=ens), contrib_ice_albedo_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Reds')
    ax.scatter(1e-16*ds_sia_avg['sia_PIP'].sel(ensemble=ens), contrib_sw_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Greens')
    ax.scatter(1e-16*ds_sia_avg['sia_PIP'].sel(ensemble=ens), contrib_total_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Greys')


# Axis formatting
ax.tick_params(which='both', labelcolor='k', labelsize=16, direction='inout', width=1, length=10, pad=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='minor', direction='in', length=5, width=1, color='k')
ax.grid(alpha=0.4)
ax.set_xlabel(r'sea ice area (million km$^2$)', fontsize=16)
ax.set_ylabel(r'Solar heating W/m$^2$', fontsize=16)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))

# Storage for polyfit data
x_all = []
y_area = []
y_alb  = []
y_sw   = []
y_tot  = []

for ens in ds_sia_avg.ensemble:
    x = 1e-16 * ds_sia_avg['sia_PIP'].sel(ensemble=ens)

    # --- scatter each variable ---
    ax.scatter(x, contrib_ice_area_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Blues')
    ax.scatter(x, contrib_ice_albedo_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Reds')
    ax.scatter(x, contrib_sw_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Greens')
    ax.scatter(x, contrib_total_MJJA.sel(ensemble=ens),
               c=ds_sia_avg.time.dt.year, s=1, cmap='Greys')

    # --- collect for polyfit ---
    x_all.append(x.values)
    y_area.append(contrib_ice_area_MJJA.sel(ensemble=ens).values)
    y_alb.append(contrib_ice_albedo_MJJA.sel(ensemble=ens).values)
    y_sw.append(contrib_sw_MJJA.sel(ensemble=ens).values)
    y_tot.append(contrib_total_MJJA.sel(ensemble=ens).values)

# --- flatten ---
x_flat  = np.concatenate(x_all)
area_f  = np.concatenate(y_area)
alb_f   = np.concatenate(y_alb)
sw_f    = np.concatenate(y_sw)
tot_f   = np.concatenate(y_tot)

def clean(x, y):
    m = np.isfinite(x) & np.isfinite(y)
    return x[m], y[m]

# Clean each dataset
x_area, area_f = clean(x_flat, area_f)
x_alb,  alb_f  = clean(x_flat, alb_f)
x_sw,   sw_f   = clean(x_flat, sw_f)
x_tot,  tot_f  = clean(x_flat, tot_f)

# --- Polyfit + plot for each variable ---
def add_fit(x, y, color, label):
    coef = np.polyfit(x, y, 3)
    poly = np.poly1d(coef)
    xfit = np.linspace(x.min(), x.max(), 300)
    ax.plot(xfit, poly(xfit), color=color, lw=2, label=label)
    print(f"{label}: slope={coef[0]:.4f}, intercept={coef[1]:.4f}")

add_fit(x_area, area_f, 'blue',  'Area contrib')
add_fit(x_alb,  alb_f,  'red',   'Albedo contrib')
add_fit(x_sw,   sw_f,   'green', 'SW contrib')
add_fit(x_tot,  tot_f,  'black', 'Total contrib')

# Format
ax.tick_params(which='both', labelcolor='k', labelsize=16,
               direction='inout', width=1, length=10, pad=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='minor', direction='in', length=5, width=1, color='k')
ax.grid(alpha=0.4)

ax.set_xlabel(r'sea ice area (million km$^2$)', fontsize=16)
ax.set_ylabel(r'Solar heating W/m$^2$', fontsize=16)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# assume your dataset is called ds
sic = ds_nsidc['cdr_seaice_conc']

# 1️⃣ Compute grid spacing (in meters)
dx = np.abs(ds_nsidc.xgrid.diff('x').mean().item())
dy = np.abs(ds_nsidc.ygrid.diff('y').mean().item())

print(f"Grid spacing: dx={dx:.1f} m, dy={dy:.1f} m")

# 2️⃣ Compute grid cell area
cell_area = dx * dy  # in m²

# 3️⃣ Convert SIC to fraction and mask invalid values
sic = sic.where((sic >= 0) & (sic <= 1))

# 4️⃣ Compute total ice-covered area (sum over spatial dims)
seaice_area = (sic * cell_area).sum(dim=['y', 'x'])

# 5️⃣ Convert to million km²
seaice_area_mkm2 = (seaice_area / 1e6) / 1e6
seaice_area_mkm2.name = 'seaice_area_million_km2'

seaice_area_mkm2
```

```python
fig, ax = plt.subplots(figsize=(6,6))

# ------------------------------------------------------------
# --- Convert all inputs to NumPy arrays immediately ---
# ------------------------------------------------------------
x = mjja_mean_sia.values
ocn = contrib_ice_area_MJJA_obs.values
ice = contrib_ice_albedo_MJJA_obs.values
sw = contrib_sw_MJJA_obs.values
tot = contrib_total_MJJA_obs.values

# --- Scatter points ---
ax.scatter(x, ocn,  c='royalblue', s=40, label='Ocean')
ax.scatter(x, ice,  c='firebrick', s=40, label='Ice')
ax.scatter(x, sw,  c='green', s=40, label='sw')
ax.scatter(x, tot,  c='gray',      s=40, label='Total')


# ============================================================
# --- 3-year running mean smoother ---
# ============================================================
def smooth(arr, w=3):
    """Centered running mean."""
    return np.convolve(arr, np.ones(w)/w, mode='same')


ocn_s   = smooth(ocn)
ice_s   = smooth(ice)
sw_s   = smooth(sw)
tot_s   = smooth(tot)

# Mask invalids (edges of running mean)
valid = (
    ~np.isnan(ocn_s)
    & ~np.isnan(ice_s)
    & ~np.isnan(sw_s)
    & ~np.isnan(tot_s)
    & ~np.isnan(x)
)

x_s    = x[valid]
ocn_s  = ocn_s[valid]
ice_s  = ice_s[valid]
sw_s  = sw_s[valid]
tot_s  = tot_s[valid]


# ============================================================
# --- Fit lines to smoothed data ---
# ============================================================
def plot_fit(x, y, color, label):
    """Fit a quadratic polynomial to smoothed data and plot."""
    x = np.asarray(x)
    y = np.asarray(y)

    coeffs = np.polyfit(x, y, deg=3)
    poly   = np.poly1d(coeffs)

    xx = np.linspace(x.min(), x.max(), 200)
    ax.plot(xx, poly(xx), color=color, lw=2.5, alpha=0.9, label=label)


plot_fit(x_s, ocn_s,  'royalblue', 'Ocean (smoothed fit)')
plot_fit(x_s, ice_s,  'firebrick', 'Ice (smoothed fit)')
plot_fit(x_s, sw_s,  'green', 'SW (smoothed fit)')
plot_fit(x_s, tot_s,  'black',     'Total (smoothed fit)')


# ============================================================
# --- Aesthetics ---
# ============================================================
ax.tick_params(which='both', labelcolor='k', labelsize=16,
               direction='inout', width=1, length=10, pad=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='minor', direction='in', length=5, width=1, color='k')

ax.grid(alpha=0.4)

ax.set_xlabel(r'Sea ice area (million km$^2$)', fontsize=16)
ax.set_ylabel(r'Solar heating (W/m$^2$)',       fontsize=16)

ax.legend(fontsize=12)

plt.show()
```

```python
# ============================================================
# Universal polynomial + slope function (any degree)
# ============================================================
def polyfit_and_slope(x, y, degree=1, npts=40):
    """
    Fit polynomial of arbitrary degree and compute its slope (derivative).
    """
    coeffs = np.polyfit(x, y, degree)   # polynomial coefficients
    poly   = np.poly1d(coeffs)          # polynomial object
    dpoly  = np.polyder(poly)           # derivative polynomial
    
    xx = np.linspace(x.min(), x.max(), npts)  # smooth x-grid
    slope = (-ds_sia_mean_obs.values/sw_dn_mean_obs.sw_dn.values)*dpoly(xx)                          # slope at each point
    
    return xx, slope, coeffs


# ============================================================
# Compute slope curves for ocn / ice / total
# (Assumes you already computed x_s, ocn_s, ice_s, total_s)
# ============================================================
degree = 3   # <-- change to whatever degree you want

x_ocn, slope_ocn,  coeffs_ocn  = polyfit_and_slope(x_s, ocn_s,   degree)
x_ice, slope_ice,  coeffs_ice  = polyfit_and_slope(x_s, ice_s,   degree)
x_sw, slope_sw,  coeffs_sw  = polyfit_and_slope(x_s, sw_s,   degree)
x_tot, slope_tot,  coeffs_tot  = polyfit_and_slope(x_s, tot_s, degree)


# ============================================================
# --------- PLOT: single panel with only slope curves ---------
# ============================================================
fig, ax = plt.subplots(figsize=(6,6))

mjja_sia_rel = mjja_mean_sia/mjja_mean_sia.sel(time=slice('1982-01-01', '2002-01-01')).mean('time')
sia_rel = np.linspace(mjja_sia_rel.min().values, mjja_sia_rel.max().values, 40)
    
ax.plot(sia_rel, slope_ocn,  color='royalblue', lw=2.5, label='Ocean slope')
ax.plot(sia_rel, slope_ice,  color='firebrick', lw=2.5, label='Ice slope')
ax.plot(sia_rel, slope_sw,  color='green', lw=2.5, label='SW slope')
ax.plot(sia_rel, slope_tot,  color='black',     lw=3.0, label='Total slope')

ax.set_xlabel(r'Sea ice area (million km$^2$)', fontsize=16)
ax.set_ylabel(r'Local slope  $\frac{d(\mathrm{heating})}{d(\mathrm{SIA})}$', fontsize=16)

ax.tick_params(which='both', labelcolor='k', labelsize=16,
               direction='inout', width=1, length=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.grid(alpha=0.4)
ax.legend(fontsize=12)

plt.show()
```

```python
# ============================================================
# --- Function: mean slope of polynomial fit ---
# ============================================================
def mean_poly_slope(x, y, degree=3):
    """
    Fit a polynomial of given degree and return the mean slope over the domain.
    This is the secant slope of the polynomial between min(x) and max(x).
    """
    x = np.asarray(x)
    y = np.asarray(y)

    coeffs = np.polyfit(x, y, degree)
    poly   = np.poly1d(coeffs)

    xmin, xmax = x.min(), x.max()
    mean_slope = (-ds_sia_mean_obs.values/sw_dn_mean_obs.sw_dn.values)*(poly(xmax) - poly(xmin)) / (xmax - xmin)

    return mean_slope, coeffs


# ============================================================
# --- Compute mean slopes for smoothed temperature curves ---
# ============================================================
degree = 1  # degree of polynomial (adjust if needed)

mean_slope_ocn,  coeffs_ocn  = mean_poly_slope(x_s, ocn_s,   degree)
mean_slope_ice,  coeffs_ice  = mean_poly_slope(x_s, ice_s,   degree)
mean_slope_sw,  coeffs_sw  = mean_poly_slope(x_s, sw_s,   degree)
mean_slope_tot,  coeffs_tot  = mean_poly_slope(x_s, tot_s, degree)

# X-coordinate for plotting: mean SIA
x_mean = np.mean(x_s)


# ============================================================
# --------- PLOT: one mean-slope point per component ---------
# ============================================================
fig, ax = plt.subplots(figsize=(6,6))

ax.scatter([x_mean], [mean_slope_ocn],  s=140, color='royalblue', label='Ocean mean slope')
ax.scatter([x_mean], [mean_slope_ice],  s=140, color='firebrick', label='Ice mean slope')
ax.scatter([x_mean], [mean_slope_sw],  s=140, color='green', label='SW mean slope')
ax.scatter([x_mean], [mean_slope_tot],  s=170, color='black',     label='Total mean slope')

# Zero reference line
ax.axhline(0, color='gray', lw=1)

# Labels
ax.set_xlabel(r'Mean sea ice area (million km$^2$)', fontsize=16)
ax.set_ylabel(r'Mean slope  $\langle d(\mathrm{heating})/d(\mathrm{SIA}) \rangle$', fontsize=16)

# Aesthetics
ax.tick_params(which='both', labelsize=16, direction='inout', width=1, length=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.grid(alpha=0.4)
ax.legend(fontsize=12)

plt.show()
```

### Final figure

```python
months = ['may','jun','jul','aug']
mjja_months = [5,6, 7,8]

# --- Create 2x2 figure ---
fig, axes = plt.subplots(2, 2, figsize=(8,6), sharex=True, sharey=False)
axes = axes.flatten()

datasets = {
    "LENS1": {
        "x": ds_sia_LENS1,
        "vars": {
            "ice area": ice_area_l1,
            "ice albedo": ice_albedo_l1,
            "shortwave": incoming_SW_l1,
            "Total": contrib_total_l1
        },
        "scale": 1e-20,
        "sw": ds_sw_dn_LENS1
    },
    "LENS2": {
        "x": ds_sia_LENS2,
        "vars": {
            "ice area": ice_area,
            "ice albedo": ice_albedo,
            "shortwave": incoming_SW,
            "Total": contrib_total
        },
        "scale": 1e-16,
        "sw": ds_sw_dn_LENS2
    }
}

colors_scatter = {
    "ice area":  'royalblue',
    "ice albedo": 'firebrick',
    "shortwave": 'green',
    "Total": 'gray'
}

colors_deriv = {
    "ice area": 'royalblue',
    "ice albedo": 'firebrick',
    "shortwave": 'green',
    "Total": 'gray'
}

for i, (lens_name, data) in enumerate(datasets.items()):

    # --- Axis indices ---
    ax_scatter = axes[i]        # top row: scatter
    ax_deriv   = axes[i+2]      # bottom row: derivatives

    x_ds = data["x"]
    y_vars = data["vars"]
    scale_factor = data["scale"]
    ds_sw_dn = data["sw"]

    # --- Scatter plot: MJJA means ---
    contrib_vars = {
        "ice area": y_vars["ice area"],
        "ice albedo": y_vars["ice albedo"],
        "shortwave": y_vars["shortwave"],
        "Total": y_vars["Total"]
    }

    # Compute MJJA yearly mean of SIA
    sia_mjja_list = []
    for m in months:
        da = x_ds[f'sia_{m}']
        da_yearly = da.groupby('time.year').mean('time', skipna=True)
        sia_mjja_list.append(da_yearly)
    sia_mjja = xr.concat(sia_mjja_list, dim='month').mean('month', skipna=True)

    sia_mjja_anom = sia_mjja.groupby('year').mean('year') - sia_mjja.sel(year=slice('1982','2002')).groupby('year').mean('year').mean('year')

    sia_mjja_rel = sia_mjja.groupby('year').mean('year')/ sia_mjja.sel(year=slice('1982','2002')).groupby('year').mean('year').mean('year')
    ensemble_size = sia_mjja.sizes['ensemble']
    years = sia_mjja['year'].values
    norm = mcolors.Normalize(vmin=years.min(), vmax=years.max())

    # --- Plot model contributions ---
    for contrib_name, contrib_data in contrib_vars.items():
        y_mjja_list = []
        for m in months:
            da = contrib_data.sel(time=contrib_data['time'].dt.month.isin(mjja_months))
            da_yearly = da.resample(time="1YE").mean('time', skipna=True)
            y_mjja_list.append(da_yearly)
        y_mjja = xr.concat(y_mjja_list, dim='month').mean('month', skipna=True)

        # x_vals = (scale_factor * sia_mjja_anom.values.flatten())
        x_vals = (sia_mjja_rel.values.flatten())
        y_vals = y_mjja.values.flatten()
        
        if lens_name =='LENS1':
            ax_scatter.scatter(x_vals, y_vals, color=colors_scatter[contrib_name], 
                           alpha=0.06, s=20, marker='x', 
                           label=contrib_name, rasterized=True)
                    
        if lens_name =='LENS2':
            ax_scatter.scatter(x_vals, y_vals, color=colors_scatter[contrib_name], 
                           alpha=0.03, s=20, marker='x', 
                           label=contrib_name, rasterized=True)

    # --- Overlay OBS data (only on top of LENS1 or LENS2) ---
    if lens_name == "LENS1":  # use one subplot to show observations
        ax_scatter.scatter(mjja_mean_sia_rel, contrib_ocn_MJJA_obs, c='royalblue', s=20, alpha=0.8, edgecolor='k', label='Obs Ice Area')
        ax_scatter.scatter(mjja_mean_sia_rel, contrib_ice_MJJA_obs, c='firebrick', s=20, alpha=0.8, edgecolor='k', label='Obs Albedo')
        ax_scatter.scatter(mjja_mean_sia_rel, contrib_total_MJJA_obs,c='gray', s=20, alpha=0.8, edgecolor='k', label='Obs Total')
        ax_scatter.scatter(mjja_mean_sia_rel, contrib_sw_MJJA_obs, c='green', s=20, alpha=0.8, edgecolor='k', label='Obs SW'  )
        # --- POLYFIT FOR EACH CONTRIBUTION ---
        

        # --- Overlay OBS data (only on top of LENS1 or LENS2) ---
    if lens_name == "LENS2":  # use one subplot to show observations
        ax_scatter.scatter(
            mjja_mean_sia_rel, contrib_ice_area_MJJA_obs,
            c='royalblue', 
            s=20, alpha=0.8, edgecolor='k', label='Obs Ice Area'
        )
        ax_scatter.scatter(
            mjja_mean_sia_rel, contrib_ice_albedo_MJJA_obs,
            c='firebrick',
            s=20, alpha=0.8, edgecolor='k', label='Obs Albedo'
            
        )

        ax_scatter.scatter(
            mjja_mean_sia_rel, contrib_sw_MJJA_obs,
            c='green',
            s=20, alpha=0.8, edgecolor='k', label='Obs SW'
        )
                
        ax_scatter.scatter(
            mjja_mean_sia_rel, contrib_total_MJJA_obs,
            c='gray',
            s=20, alpha=0.8, edgecolor='k', label='Obs Total'
        )

    # --- Vertical lines for 2014 and 2060 ---
    for year_mark, line_color in zip([2014, 2060], ["dimgray", "dimgray"]):
        if year_mark in years:
            idx = np.where(years == year_mark)[0][0]
            # sia_mean = (scale_factor * sia_mjja_anom.isel(year=idx)).mean().item()
            sia_mean = (sia_mjja_rel.isel(year=idx)).mean().item()
            ax_scatter.axvline(sia_mean, color=line_color, linestyle='--', linewidth=1.4, alpha=0.9)

    # Axis formatting
    ax_scatter.tick_params(which='both', labelsize=14)
    ax_scatter.xaxis.set_minor_locator(AutoMinorLocator())
    ax_scatter.yaxis.set_minor_locator(AutoMinorLocator())
    ax_scatter.axhline(0, color='k', lw=0.4)

    # --- Derivative / effective albedo plot ---
    # Compute MJJA SW baseline (1920-1950)
    sw_list = []
    for m_idx, m in enumerate(months):
        sw_month = ds_sw_dn['sw_dn'].sel(time=ds_sw_dn['time'].dt.month==(5+m_idx)).resample(time="1YE").mean()
        sw_list.append(sw_month.sel(time=slice('1982-01-01','2002-01-01')).mean(dim=['ensemble','time']))
    sw_const = xr.concat(sw_list, dim='month').mean('month').values

    # SIA baseline
    sia_base_list = []
    for m in months:
        da = x_ds[f'sia_{m}'].dropna(dim='time')
        sia_base_list.append(da.sel(time=slice('1982-01-01','2002-01-01')).mean(dim=['ensemble','time']))
    sia_const = xr.concat(sia_base_list, dim='month').mean('month').values

    for label, y_da in y_vars.items():
        derivs = []

        y_mjja_list = []
        for m in months:
            da = y_da.sel(time=y_da['time'].dt.month.isin(mjja_months))
            da_yearly = da.resample(time="1YE").mean('time', skipna=True)
            y_mjja_list.append(da_yearly)
        y_mjja = xr.concat(y_mjja_list, dim='month').mean('month', skipna=True)

        for ens in range(y_da.sizes['ensemble']):
            x = (scale_factor * sia_mjja_anom.isel(ensemble=ens).values)
            x_rel = (sia_mjja_rel.isel(ensemble=ens).values)
            y = y_mjja.isel(ensemble=ens).values
            mask = ~np.isnan(x) & ~np.isnan(y)
            if mask.sum() < 4:
                continue
            coeffs = np.polyfit(x[mask], y[mask], deg=3)
            x_sorted = np.linspace(x[mask].min(), x[mask].max(), 200)
            x_rel_sorted = np.linspace(x_rel[mask].min(), x_rel[mask].max(), 200)
            dydx = -(scale_factor * sia_const / sw_const) * np.polyval(np.polyder(coeffs), x_sorted)
            # dydx = (scale_factor*sia_const / sw_const) * np.polyval(np.polyder(coeffs), x_sorted)
            derivs.append(dydx)

        if derivs:
            derivs = np.array(derivs)
            dydx_mean = derivs.mean(axis=0)
            dydx_std = derivs.std(axis=0)
            ax_deriv.plot((x_rel_sorted), dydx_mean, color=colors_deriv[label], lw=2.5, label=f"{label} "+r"$\alpha_e$")
            ax_deriv.fill_between((x_rel_sorted), dydx_mean-2*dydx_std, dydx_mean+2*dydx_std, color=colors_deriv[label], alpha=0.2)

      
        # Vertical lines for 2014 and 2060
        for year_mark, line_color in zip([2014, 2060], ["dimgray", "dimgray"]):
            if year_mark in years:
                idx = np.where(years == year_mark)[0][0]
                # sia_mean = (scale_factor * sia_mjja_anom.isel(year=idx)).mean().item()
                # sia_mean = (sia_mjja_rel.isel(year=idx)).mean().item()
                sia_mean = (sia_mjja_rel.isel(year=idx)).mean().item()
                ax_deriv.axvline(sia_mean, color=line_color, linestyle='--', linewidth=1.4, alpha=0.9, zorder=5)

                # --- After plotting model + OBS scatter ---
    if lens_name in ['LENS1', 'LENS2']:
        # Plot OBS mean slope points
        ax_deriv.scatter([np.mean(mjja_mean_sia_rel)], [mean_slope_ocn], s=100, marker='*', color='royalblue')
        ax_deriv.scatter([np.mean(mjja_mean_sia_rel)], [mean_slope_ice], s=100, marker='*', color='firebrick')
        ax_deriv.scatter([np.mean(mjja_mean_sia_rel)], [mean_slope_sw], s=100, marker='*', color='green')
        ax_deriv.scatter([np.mean(mjja_mean_sia_rel)], [mean_slope_tot], s=100, marker='*', color='gray', zorder=5)
    


    ax_deriv.tick_params(which='both', labelsize=14)
    ax_deriv.xaxis.set_minor_locator(AutoMinorLocator())
    ax_deriv.yaxis.set_minor_locator(AutoMinorLocator())
    ax_deriv.set_ylim(-0.3, 0.7)
    # ax_deriv.set_xlim(4.5, 9)

# --- Shared labels ---
fig.text(0.55, 0.02, r'SIA/SIA$_{1982-2002}$', ha='center', fontsize=16)
axes[0].set_ylabel(r'W/m$^2$',  fontsize=18)
axes[2].set_ylabel(r'$\alpha^e$',  fontsize=18)

# # Legend for models + obs

legend_label = {
    'Ice area': 'royalblue',
    'Ice albedo': 'firebrick',
    'Shortwave': 'Green',
    'Sum': 'k'
}

handles_scatter = [
    Line2D([0], [0], marker='o', color='w',
           markerfacecolor=c, markersize=8, label=l)
    for l, c in legend_label.items()
]

# handles_obs = [Line2D([0], [0], marker='o', color='w', markerfacecolor='k', markersize=8, label='Obs')]
fig.legend(handles=handles_scatter, loc='upper center', ncol=4, fontsize=12, frameon=False)

# Add subplot labels
subplot_labels = ['a', 'b', 'c', 'd']
for ax, label in zip(axes, subplot_labels):
    ax.text(0.95, 0.95, f'({label})', transform=ax.transAxes,
            fontsize=14, fontweight='bold', va='top', ha='right')

# # Right-side bold LENS labels
# Add label next to ax[0] (LENS1)
axes[0].text(
    0.4,       # a bit outside the right border
    1.05,        # center vertically inside axis
    r"\textbf{LENS1}",
    fontsize=14,
    fontweight='bold',
    color='black',
    ha='left',
    va='center',
    transform=axes[0].transAxes
)

# Add label next to ax[1] (LENS2)
axes[1].text(
    0.35,
    1.05,
    r"\textbf{LENS2}",
    fontsize=14,
    fontweight='bold',
    color='black',
    ha='left',
    va='center',
    transform=axes[1].transAxes
)

axes[0].set_ylim(-25,80)
axes[1].set_ylim(-25,80)
axes[1].set_xlim(0.4,1)

for ax in axes:
    ax.tick_params(which='both', labelsize=14, direction='inout', width=1, length=8, pad=6)
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(which='minor', direction='in', length=4, width=1, color='k')
    ax.grid(alpha=0.2)

plt.tight_layout(rect=[0.05,0.05,0.95,0.95])
# plt.savefig('../figs/eff_albedos_with_obs_sia_rel.pdf', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# ===============================
# ANALYSIS / DATA PREP CELL
# ===============================

months = ['may', 'jun', 'jul', 'aug']
mjja_months = [5, 6, 7, 8]

def compute_mjja_mean(da):
    mjja = []
    for m in months:
        da_year = da.groupby('time.year').mean('time', skipna=True)
        mjja.append(da_year)
    return xr.concat(mjja, dim='month').mean('month', skipna=True)


def compute_mjja_resampled(da):
    """
    Returns MJJA mean with dims (ensemble, year)
    """
    tmp = da.sel(time=da.time.dt.month.isin(mjja_months))
    tmp = tmp.resample(time="1YE").mean("time", skipna=True)

    # Convert time → year coordinate and drop time
    tmp = tmp.assign_coords(year=tmp.time.dt.year)
    tmp = tmp.swap_dims({"time": "year"}).drop_vars("time")

    return tmp



analysis = {}

datasets = {
    "LENS1": {
        "sia": ds_sia_LENS1,
        "vars": {
            "Ice area": ice_area_l1,
            "Ice albedo": ice_albedo_l1,
            "Shortwave": incoming_SW_l1,
            "Sum": contrib_total_l1
        },
        "sw": ds_sw_dn_LENS1,
        "scale": 1e-20
    },
    "LENS2": {
        "sia": ds_sia_LENS2,
        "vars": {
            "Ice area": ice_area,
            "Ice albedo": ice_albedo,
            "Shortwave": incoming_SW,
            "Sum": contrib_total
        },
        "sw": ds_sw_dn_LENS2,
        "scale": 1e-16
    }
}

for name, d in datasets.items():

    sia_ds = d["sia"]
    scale = d["scale"]

    # --- MJJA SIA ---
    sia_mjja = compute_mjja_mean(
        xr.concat([sia_ds[f'sia_{m}'] for m in months], dim='month')
    )

    sia_base = sia_mjja.sel(year=slice('1982', '2002')).mean(['year'])
    sia_anom = (sia_mjja - sia_mjja.sel(year=slice('1982', '2002')).mean('year'))
    sia_rel  = sia_mjja / sia_base

    years = sia_mjja['year'].values

    # --- Contributions ---
    contrib = {}
    for label, da in d["vars"].items():
        y_mjja = compute_mjja_resampled(da)
        contrib[label] = xr.Dataset({"x": sia_rel,
                                     "y": y_mjja
                                        })

    # --- SW & SIA constants for derivative ---
    sw_list = []
    for m_idx in range(4):
        sw_m = d["sw"]['sw_dn'].sel(
            time=d["sw"]['time'].dt.month == (5 + m_idx)
        ).resample(time='1YE').mean()
        sw_list.append(sw_m.sel(time=slice('1982', '2002')).mean(['time']))
    sw_const = xr.concat(sw_list, dim='month').mean('month').values

    sia_const = sia_base.values

    # --- Effective albedo derivatives ---
    derivs = {}

    for label, da in d["vars"].items():

        y_mjja = compute_mjja_resampled(da)
        curves = []
        
        dydx_list = []
        ens_list = []

        for ens in range(da.sizes['ensemble']):
            x = scale * sia_anom.isel(ensemble=ens).values
            x_rel = sia_rel.isel(ensemble=ens).values
            y = y_mjja.isel(ensemble=ens).values

            mask = ~np.isnan(x) & ~np.isnan(y)
            if mask.sum() < 4:
                continue

            coeffs = np.polyfit(x[mask], y[mask], 3)
            x_sorted = np.linspace(x_rel[mask].min(), x_rel[mask].max(), 200)
            dydx = -(scale * sia_const[ens] / sw_const[ens]) * np.polyval(
                np.polyder(coeffs),
                np.linspace(x[mask].min(), x[mask].max(), 200)
            )
            
            dydx_list.append(dydx)
            ens_list.append(ens)

        if dydx_list:
            dydx_arr = xr.DataArray(
                np.stack(dydx_list, axis=0),
                dims=["ensemble", "x"],
                coords={
                    "ensemble": ens_list,
                    "x": x_sorted
                }
            )
        
            derivs[label] = xr.Dataset(
                {
                    "dydx": dydx_arr,
                    "mean": dydx_arr.mean("ensemble"),
                    "std":  dydx_arr.std("ensemble")
                }
            )


    analysis[name] = {
    "scatter": contrib,        # dict of xr.Dataset (ensemble, year)
    "deriv": derivs,           # dict of xr.Dataset (ensemble, x)
    "years": years,
    "sia_rel": sia_rel         # (ensemble, year)
    }
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.ticker import AutoMinorLocator
import xarray as xr
import pandas as pd

# -------------------------------
# Helper function: split label into two lines
# -------------------------------
def two_line_labels(labels):
    return [l.replace(" ", "\n") for l in labels]

# ===============================
# Figure setup (2 x 3)
# ===============================
fig, axes = plt.subplots(
    2, 3,
    figsize=(10,6),
    sharex='col',
    gridspec_kw={"width_ratios": [1, 1, 1.5]}
)

ax_scatter = axes[0, 0:2]
ax_deriv   = axes[1, 0:2]
ax_violin  = axes[:, 2]  # single column for violin plots

lens_names = ["LENS1", "LENS2"]

colors = {
    "Ice area": "royalblue",
    "Ice albedo": "firebrick",
    "Shortwave": "green",
    "Sum": "gray"
}

period = slice(1982, 2020)

# ===============================
# Loop over LENS for scatter + deriv
# ===============================
for i, lens in enumerate(lens_names):
    ax_s = ax_scatter[i]
    ax_d = ax_deriv[i]

    scatter_data = analysis[lens]["scatter"]
    deriv_data   = analysis[lens]["deriv"]
    sia_rel      = analysis[lens]["sia_rel"]
    years        = analysis[lens]["years"]

    # -------------------------------
    # Scatter plots
    # -------------------------------
    for label, ds in scatter_data.items():
        ax_s.scatter(
            ds["x"].values,
            ds["y"].values,
            color=colors[label],
            alpha=0.05,
            s=18,
            marker="x",
            rasterized=True
        )
    ax_s.axhline(0, color="k", lw=0.4)

    # -------------------------------
    # Vertical year markers
    # -------------------------------
    for year_mark in [2014, 2060]:
        if year_mark in years:
            idx  = np.where(years == year_mark)[0][0]
            xval = sia_rel.isel(year=idx).mean("ensemble").item()
            ax_s.axvline(xval, color="dimgray", ls="--", lw=1.4)
            ax_d.axvline(xval, color="dimgray", ls="--", lw=1.4)

    # -------------------------------
    # Effective albedo curves
    # -------------------------------
    for label, ds in deriv_data.items():
        x    = ds["mean"]["x"].values
        mean = ds["mean"].values
        std  = ds["std"].values

        ax_d.plot(x, mean, color=colors[label], lw=2.5)
        ax_d.fill_between(x, mean - 2*std, mean + 2*std,
                          color=colors[label], alpha=0.2)

    ax_s.set_ylim(-25, 80)
    ax_d.set_ylim(-0.3, 0.7)
    ax_d.set_xlim(0.4, 1.0)

# ===============================
# Prepare data for violin plots (Lens1 + Lens2 side by side)
# ===============================
# Contribution violins
ax = ax_violin[0]

scatter_labels = list(analysis[lens_names[0]]["scatter"].keys())
n_labels = len(scatter_labels)
n_lens   = len(lens_names)
width = 0.8

positions = []
data = []

for i, label in enumerate(scatter_labels):
    for j, lens in enumerate(lens_names):
        da = analysis[lens]["scatter"][label]["y"].sel(year=period).mean("year")
        data.append(da.values)
        positions.append(i*n_lens + j)

# Create violin plots
vp = ax.violinplot(
    data,
    positions=positions,
    widths=0.7,
    showmedians=True,
    showextrema=True
)

# Violin bodies
for body in vp['bodies']:
    body.set_facecolor(colors[label])  # if you color-fill
    body.set_edgecolor('black')
    body.set_linewidth(1.2)
    body.set_alpha(0.85)

# Lines (median, min/max, bars)
for part in ('cmedians', 'cmins', 'cmaxes', 'cbars'):
    if part in vp:
        vp[part].set_color('black')
        vp[part].set_linewidth(1.2)

# Color violins
for idx, patch in enumerate(vp['bodies']):
    label = scatter_labels[idx // n_lens]
    patch.set_facecolor(colors[label])
    patch.set_alpha(0.85)

# OBS points (mean over 1982–2020)
period_dt = slice(pd.Timestamp("1982-01-01"), pd.Timestamp("2020-12-31"))
obs_means_contrib = [
    contrib_ice_area_MJJA_obs.sel(time=period_dt).mean().item(),
    contrib_ice_albedo_MJJA_obs.sel(time=period_dt).mean().item(),
    contrib_sw_MJJA_obs.sel(time=period_dt).mean().item(),
    contrib_total_MJJA_obs.sel(time=period_dt).mean().item()
]

for i, label in enumerate(scatter_labels):
    xpos_obs = i*n_lens + 0.5
    ax.scatter(xpos_obs, obs_means_contrib[i], s=100, marker='*',
               color=colors[label], edgecolor='k', zorder=5)

ax.set_xticks([i*n_lens + 0.5 for i in range(n_labels)])
xticks = ax.set_xticklabels(two_line_labels(scatter_labels))
for tick_label, label in zip(xticks, scatter_labels):
    tick_label.set_color(colors[label])
    tick_label.set_fontsize(12)
# ax.set_ylabel(r"W/m$^{2}$", fontsize=14)


# Effective albedo violins
ax2 = ax_violin[1]

deriv_labels = list(analysis[lens_names[0]]["deriv"].keys())
positions = []
data = []

for i, label in enumerate(deriv_labels):
    for j, lens in enumerate(lens_names):
        deriv_data = analysis[lens]["deriv"]
        sia_p = analysis[lens]["sia_rel"].sel(year=period)
        sia_min = sia_p.min("year")
        sia_max = sia_p.max("year")
        dydx = deriv_data[label]["dydx"]
        x = dydx["x"]

        alpha_ens = []
        for ens in range(39):
            mask = (x >= sia_min.sel(ensemble=ens)) & (x <= sia_max.sel(ensemble=ens))
            if mask.sum() < 5:
                continue
            alpha_ens.append(dydx.sel(ensemble=ens).where(mask, drop=True).mean("x"))

        if len(alpha_ens) == 0:
            continue
        alpha_ens = xr.concat(alpha_ens, dim="ensemble")
        data.append(alpha_ens.values)
        positions.append(i*n_lens + j)

vp2 = ax2.violinplot(data, positions=positions, widths=0.7, showmedians=True, showextrema=True)


# Violin bodies
for body in vp2['bodies']:
    body.set_facecolor(colors[label])  # if you color-fill
    body.set_edgecolor('black')
    body.set_linewidth(1.2)
    body.set_alpha(0.85)

# Lines (median, min/max, bars)
for part in ('cmedians', 'cmins', 'cmaxes', 'cbars'):
    if part in vp2:
        vp2[part].set_color('black')
        vp2[part].set_linewidth(1.2)

# Color violins
for idx, patch in enumerate(vp2['bodies']):
    label = deriv_labels[idx // n_lens]
    patch.set_facecolor(colors[label])
    patch.set_alpha(0.85)

# OBS points centered between LENS1 and LENS2
obs_values = {
    "Ice area": mean_slope_ocn,
    "Ice albedo": mean_slope_ice,
    "Shortwave": mean_slope_sw,
    "Sum": mean_slope_tot
}

for i, label in enumerate(obs_values.keys()):
    xpos = i*n_lens + 0.5
    ax2.scatter(xpos, obs_values[label], marker='*', s=100,
                color=colors[label], edgecolor='k', zorder=5)

ax2.set_xticks([i*n_lens + 0.5 for i in range(len(deriv_labels))])
xticks2 = ax2.set_xticklabels(two_line_labels(deriv_labels))
for tick_label, label in zip(xticks2, deriv_labels):
    tick_label.set_color(colors[label])
    tick_label.set_fontsize(12)
# ax2.set_ylabel(r"$\alpha^e$", fontsize=14)
ax2.set_ylim(-0.3, 0.7)

# ===============================
# Shared labels & legend
# ===============================
axes[0,0].set_ylabel(r"W/m$^{2}$", fontsize=20)
axes[1,0].set_ylabel(r"$\alpha^e$", fontsize=20)
axes[1,0].set_xlabel(r"SIA / SIA$_{1982–2002}$", fontsize=16)
axes[1,1].set_xlabel(r"SIA / SIA$_{1982–2002}$", fontsize=16)

handles = [Line2D([0],[0], color=c, lw=3, label=l) for l,c in colors.items()]
fig.legend(handles=handles, loc="upper center", ncol=4, fontsize=14, frameon=False)

# -------------------------------
# Subplot labels
# -------------------------------
subplot_labels = ['a','b','c','d','e','f']

# Flatten axes and assign labels
for ax, label in zip(axes.flatten(), subplot_labels):
    ax.text(
        0.97, 0.95, f"({label})",
        transform=ax.transAxes,
        fontsize=14,
        fontweight="bold",
        va="top",
        ha="right"
    )


# -------------------------------
# Final formatting
# -------------------------------
for ax_row in axes:
    for ax in ax_row:
        ax.tick_params(direction="inout", length=8, width=1, labelsize=16)
        ax.tick_params(which="minor", direction="in", length=4, width=1)
        ax.xaxis.set_minor_locator(AutoMinorLocator())
        ax.yaxis.set_minor_locator(AutoMinorLocator())
        ax.grid(alpha=0.2)

# Reduce spacing between subplots
plt.subplots_adjust(
    left=0.03,    # left edge
    right=0.97,   # right edge
    bottom=0.03,  # bottom edge
    top=0.84 ,    # top edge
    wspace=0.3,  # horizontal space between columns
    hspace=0.25   # vertical space between rows
)

axes.flatten()[0].text(
    0.4,       # a bit outside the right border
    1.07,        # center vertically inside axis
    r"\textbf{LENS1}",
    fontsize=14,
    fontweight='bold',
    color='black',
    ha='left',
    va='center',
    transform=axes.flatten()[0].transAxes
)

# Add label next to ax[1] (LENS2)
axes.flatten()[1].text(
    0.35,
    1.07,
    r"\textbf{LENS2}",
    fontsize=14,
    fontweight='bold',
    color='black',
    ha='left',
    va='center',
    transform=axes.flatten()[1].transAxes
)

plt.savefig('../figs/eff_albedos_with_obs_sia_rel_violin_plot.pdf', bbox_inches='tight', dpi=300)
```
